```python
data= data.drop(['Comments'],axis=1)
```

```python
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data['Timestamp'], data['GHI'], color='blue')
plt.title('Global Horizontal Irradiance (GHI) over Time')
plt.xlabel('Timestamp')
plt.ylabel('GHI')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data['Timestamp'], data['DNI'], color='red')
plt.title('Direct Normal Irradiance (DNI) over Time')
plt.xlabel('Timestamp')
plt.ylabel('DNI')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data['Timestamp'], data['DHI'], color='green')
plt.title('Diffuse Horizontal Irradiance (DHI) over Time')
plt.xlabel('Timestamp')
plt.ylabel('DHI')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data['Timestamp'], data['Tamb'], color='orange')
plt.title('Ambient Temperature (Tamb) over Time')
plt.xlabel('Timestamp')
plt.ylabel('Tamb')
plt.grid(True)
plt.show()
```

```python
# Calculate correlation matrix
correlation_matrix = data[['GHI', 'DHI', 'DNI', 'TModA', 'TModB']].corr()
```

```python
# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
#Convert 'Timestamp' column to datetime format
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
```

```python
# Plot wind speed (WS) over time
plt.figure(figsize=(12, 6))
plt.plot(data['Timestamp'], data['WS'], color='blue', label='WS')
plt.plot(data['Timestamp'], data['WSgust'], color='red', label='WSgust')
plt.plot(data['Timestamp'], data['WSstdev'], color='green', label='WSstdev')
plt.title('Wind Speed Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Wind Speed (m/s)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot wind direction (WD) over time
plt.figure(figsize=(12, 6))
plt.plot(data['Timestamp'], data['WD'], color='blue', label='WD')
plt.plot(data['Timestamp'], data['WDstdev'], color='red', label='WDstdev')
plt.title('Wind Direction Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Wind Direction (degrees)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate descriptive statistics
wind_speed_stats = data[['WS', 'WSgust', 'WSstdev']].describe()
wind_direction_stats = data[['WD', 'WDstdev']].describe()
print("Wind Speed Statistics:")
print(wind_speed_stats)
print("\nWind Direction Statistics:")
print(wind_direction_stats)
```

```python
# Select the columns for solar radiation and temperature data
solar_radiation_columns = ['GHI', 'DNI', 'DHI']
temperature_columns = ['TModA', 'TModB', 'Tamb']
```

```python
# Combine the selected columns into a single DataFrame
selected_columns = solar_radiation_columns + temperature_columns
selected_data = data[selected_columns]
```

```python
# Box plot for solar radiation data
plt.subplot(1, 2, 1)
sns.boxplot(data=selected_data[solar_radiation_columns], palette='Set2')
plt.title('Solar Radiation Data Box Plot')
plt.ylabel('Solar Radiation (W/m^2)')
```

```python
# Box plot for temperature data
plt.subplot(1, 2, 2)
sns.boxplot(data=selected_data[temperature_columns], palette='Set1')
plt.title('Temperature Data Box Plot')
plt.ylabel('Temperature (Â°C)')
```
